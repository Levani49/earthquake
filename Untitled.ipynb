```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Earthquakes_database.csv")

df.head()
```

```python
#დავალება N1
df[['year', 'month', 'day']] = df['Date'].str.split('-', expand=True)

df[['hour', 'minute', 'second']] = df['Time'].str.split(':', expand=True)

df = df.drop(['Date', 'Time'], axis=1)

print(df)
```

```python
#დავალება N2
depth_bins = [-float('inf'), 50, 200, float('inf')]
depth_labels = ['Shallow', 'Moderate', 'Deep']

df['Category'] = pd.cut(df['Depth'], bins=depth_bins, labels=depth_labels, include_lowest=True)

print(df)
```

```python
#დავალება N3
df['Year'] = pd.to_datetime(df['Depth']).dt.year

pivot_table_df = pd.pivot_table(df, values='Depth', index='Type', columns='Year', aggfunc='mean', fill_value=0)

print(pivot_table_df)
```

```python
#დავალება N3
melted_df = pd.melt(df, id_vars=['Latitude', 'Longitude', 'Type', 'Depth'], var_name='Variable', value_name='Value')

print(melted_df)
```

```python
#დავალება N4
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value1': np.random.randint(1, 100, 5),
    'Value2': np.random.randint(1, 100, 5),
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.plot(df['Category'], df['Value1'], marker='o', label='Value1')
plt.plot(df['Category'], df['Value2'], marker='o', label='Value2')
plt.title('Line Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.show()

# Chart 2: Bar chart
plt.figure(figsize=(8, 5))
plt.bar(df['Category'], df['Value1'], alpha=0.7, label='Value1')
plt.bar(df['Category'], df['Value2'], alpha=0.7, label='Value2', bottom=df['Value1'])
plt.title('Stacked Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(df['Value1'], df['Value2'])
plt.title('Scatter Chart')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.show()

plt.figure(figsize=(8, 5))
plt.pie(df['Value1'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

```python
#დავალება N5
data2 = {
    'StartDate': pd.date_range(start='2023-01-01', end='2023-01-10'),
    'Value': np.random.randint(1, 100, 10),
}

df = pd.DataFrame(data2)

df['StartYear'] = df['StartDate'].dt.year
df['StartMonth'] = df['StartDate'].dt.month
df['StartDay'] = df['StartDate'].dt.day

df['TimeDiff'] = df['StartDate'].diff()

# Optional: Resample (assuming you have time-series data)
df.set_index('StartDate', inplace=True)
resampled_df = df.resample('D').sum()  # Resample to daily frequency, summing values

# Display the modified DataFrame
print(df)
print(resampled_df)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Value'], label='Original Data')
plt.plot(resampled_df.index, resampled_df['Value'], label='Resampled Data', linestyle='--', marker='o')
plt.title('Original vs Resampled Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
